```python
country=314
mask=S==country
img=S.copy()
img[mask]=0
plt.subplot(121)
plt.imshow(img)
plt.subplot(122)
plt.imshow(mask)
plt.show()
```

```python
hist=np.zeros(np.max(S)+1)

for i in range(S.shape[0]):
    for j in range(S.shape[1]):
        if S[i,j]:
            hist[S[i,j]]+=1
```

```python
plt.close
area=hist
area[144]=0 #russia
area[126]=0 #ukraine
area[22]=0#turkey
area[294]=0 #norway
area[7]=0#algeria
area[310]=0#finnland
area[232]=0#belarus
area[8]=0#maroko

area[315]=0#greenland
plt.plot(area)
plt.show()
```
